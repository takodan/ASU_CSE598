## M7U1
1. 選擇問題
    1. 缺乏足夠資源來嘗試 (信息過載)
    2. 防禦性決策 (水桶效應)：選擇一個不是最好，但是沒有明顯缺點的選擇
2. 推薦系統的挑戰
    1. 冷啟動問題
        1. 新的選項或是用戶，缺乏有用信息來進行推薦
    2. 數據稀疏
        1. 得到的數據過於分化，無法有效的進行推薦
    3. 網路攻擊
        1. 虛假評論、信息
    4. 隱私
        1. 推薦的同時保護隱私
    5. 可解釋性
        1. 推薦系統通常缺乏可解釋性
3. 基於內容的推薦
    1. 假設：用戶興趣和項目描述應箱匹配
    2. 目標：找到用戶和項目之間的相似性
    3. 步驟
        1. 描述項目
        2. 創建用戶資料：
        3. 將項目和用戶資料進行比較：使用關鍵字向量 (例如TF-IDF) 計算餘弦相似性
        4. 持續自動更新個人資料
4. CF協同過濾推薦
    1. 使用多個不同信息 (例如購買歷史或用戶評分) 來形成用戶項目矩陣，以矩陣進行推薦
    2. 基於記憶的協同過濾：以過去的用負項目矩陣進行推薦
        1. 基於用戶：找尋相似的歷史用戶來預測目標用戶的分數
            - 例子：U1和U2在過去三個項目的評分相似，U1和U2在第四個項目的分數也可能相似
        2. 基於項目：尋找相似的歷史項目來預測目標用戶的分數
            - 例子：每個用戶對於I3和I4的評分相似，U1對I4的評分可能會相似
        3. 步驟
            1. 計算用戶相似性或項目相似性
            2. 找套相似用戶或項目作為標準進行評分，可以找k個相似的用戶或項目
            3. 根據分數推薦項目
            4. 餘弦相似性或皮爾森積差相關
        4. 實際上大多使用基於項目方法，因為項目分數較好取得且穩定
        5. 計算例子：13:47
    3. 基於模型的協同過濾：以數學模型 (SVD) 假設矩陣，並以此進行推薦
5. 其他推薦算法
    1. 群組推薦
        1. 聚合策略
            1. 最大化平均滿意度：取平均最喜歡的
            2. 最小化厭惡：排除大家最討厭的
            3. 最大化喜歡：只取其中一個最喜歡的
        2. 計算例子：05:48
    2. 混合算法
        1. pipeline管道混合：以一個算法的輸出作為另一個算法的輸入，進行多輪計算
            1. 可以層層篩選
            2. 或是將一個輸出作為下輪計算的一個參數
        2. 並行混合：將多個算法的輸出統整成結果
            1. 混合
            2. 加權
            3. 多數決

## M7U2 評價推薦系統
1. 準確性指標
    1. 預測的準確性
    2. 推薦的相關性
    3. 推薦排序的差異
2. 準確性
    1. MAE平均絕對誤差：(預測－實際)絕對值的和／用戶數
    2. NAME均值平均絕對誤差：(MAE)／(可能最高分－可能最低分)
    3. RMSE均方根誤差：((預測－實際)平方的和/用戶數)開根號
3. 相關性
    1. 準確率：相關且被選中／全部被選中
    2. 招回率：相關且被選中／全部有相關
    3. F：P\*R*2／P＋R
4. 排序 
    1. Spearman's rank correlation coefficient排序相關係數
        - (預測名次-實際名次)平方和*6／(n三次方)－n
    2. Kendall rank correlation coefficient排序相關係數
        1. 找出所有組合
        2. 組合的箱後順序是否一致
            - 例如：i1和i2預測是1、2名，實際是1、4名，仍然可以算做一致
        3. (一致組合－不一致組合)／所有組合

## M7U3 社交媒體推薦
1. 影片推薦
    1. 關聯規則挖掘
        1. 試著計算所有用戶看完A內容後看B內容的機率
        2. 看完A內容後看B內容的人數／看完A內容的總人數
        3. 依據此機率推薦內容
    2. 共同訪問
        1. 看完A內容後看B內容的人數/歸因化因子
        2. 歸因化可以是(看完A內容後看B內容的人數*看完A內容後"不"看B內容的人數)
2. 新聞推薦
    1. 時效性最為重要
    2. 推薦不完全相似的內容反而可能使用戶更感興趣
3. 貼文推薦
    1. 需要處理作者和文章的關係
        1. 有些人可能會偏好某個作者的內容，而非其他相似但不知名作者的內容
        2. 使用不同的文檔模型
            1. 大文檔：將同作者的內容視為一個整體文檔
            2. 小文檔：每篇內容最為一個文檔
    2. 關鍵字查詢
        1. 貼文的查詢有時會是一個比較模糊或廣泛的概念，有時並不會直交接連接到用戶想要的內容
            - 例如用戶想要找一個最近很熱門遊戲的內容，但是不知道遊戲名字
        2. QE查詢擴展
            1. 例如利用維基百科的內容逕行擴展，再來搜索
4. 基於標籤
    1. 標籤是由社區用戶非強制的產生
    2. 用戶可以產生許多不同標籤，但只有少數標籤會被大量重複使用
    3. 最終有效的標籤可以簡化推薦系統，直接使用標籤作為推薦


## M7U3 推薦的解釋性
1. 置信度指標
    1. 透明度：提供運作信息
    2. 有效性： 提供運作方式解釋
    3. 可信度：推薦的準確率
    4. 可理解性：和用戶已知概念相連來使其理解
    5. 教育：提供用戶更深入的內容
    6. 說服力：有說服力的解釋
    7. 效力：推薦對於用戶最終決策的影響
    8. 效率：增加用戶決策效率
    9. 滿意度：用戶整體滿意度
    10. 相關性：如需要用戶提供額外信息，需要證明和用戶當前行為有關
2. 解釋推薦
    1. 最近鄰解釋：其他的購買A的用戶也購買B
    2. 內容解釋：此項目包含你有感興趣的主題A
    3. 社會網路解釋：你的朋友A和B也喜歡此項目
3. 提供反饋機會
