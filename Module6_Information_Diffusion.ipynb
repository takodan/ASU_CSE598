    1. Medium媒介
    2. Receivers接受者
    3. Senders傳播者
2. 傳播類型
    1. 顯式網路
        1. 全局信息可獲得：羊群效應
        2. 部分信息可獲得：信息級聯
    2. 隱式網路：流行病傳播

## M6U2L1 Herd Effect
1. 羊群效應
    1. 可觀察網路
    2. 可觀多數信息，
    3. 也稱從眾效應
    4. 組成
        1. 個人之間的聯席
        2. 轉移行為或觀察行為的方法
    5. 一般而言發生羊群效應的網路接近完全圖，節點可以觀察到多數其他節點信息
2. 示例
    1. 網路拍賣
        1. 可觀察到拍賣者和出價者的販售和競價行為
        2. 可以看到所有參與者得歷史信息，如信譽
        3. 因為以上行為和信息，可能會使得他人跟進競價同件商品
    2. Note on the Drawing power of crowds of different size
        1. 社會心裡學實驗
        2. 不同數量的人在街頭看天空，會影響不同數量的人一起看天空
    3. Asch conformity experiments
        1. 社會心裡學實驗
        2. 進行簡單的線條比長短測驗，讓其助手給出錯誤答案，部分人也會跟著給出錯誤答案
    4. 電梯從眾實驗
    5. 容器實驗
        1. 箱子理有三顆珠子，可能是紅紅藍或藍藍紅
        2. 請學生取出一顆珠子看顏色，然後在黑板上紀錄預測是紅色多或是藍色多
        3. 學生均學過機率，後面的學生可以看到前面學生的預測
        4. 根據貝葉斯法則(條件機率)，抽到什麼顏色代表抽到的顏色是多數的比率較高
        4. 結果顯示後面學生會受前面影響學生預測影響，而非完全依照取出顏色判斷
3. 干預方式
    1. 透過徹底公開信息或是增加信息的方式來干預
        1. 例如在實驗中，公開給出錯誤訊息的人是演員、助手
4. Knowledge Check補充
    1. 一群人未經過計劃做出相同行為也屬於羊群效應
    2. 阻止所有信息也可以阻止羊群效應

## M6U3L1 Information Cascades
1. 信息級聯
    1. 可觀察網路
    2. 僅局部信息可用
    3. 相比羊群效應，只能觀察到直接鄰居
    4. 網路是有向圖，接近數狀結構
    5. 節點是二進制的，活躍或不活躍，活躍節點可以激或其他鄰近節點
    6. 有多種級聯模型
2. 獨立級聯模型
    1. 傳播者：活躍節點
    2. 接收者：被激活節點
3. 簡單來說就是傳銷，一傳十，十傳百
4. 干預
    1. 最大化聯集傳播
        1. 找到邊多且易傳播的節點
        2. 假設
            1. 給定參數(節點數)k，求出有k個活躍節點的節點集S，最大化活躍節點期望數f(S)
            2. f(S+v) >= f(S)，初始節點數越多，期望值只會越好
            3. 邊際效益遞減：一開始初始節點少，每個增加的節點對期望值的影響越大
        3. 是NP難題
        4. 可以使用貪婪算法，S從空集合開始，迭代k次
        5. 貪婪算法可以至少達到最優解的63%
    2. 降低f(S)
        1. 斷開連結
        2. 降低激活可能性(閾值)
