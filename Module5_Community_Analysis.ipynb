## M5U1L1 Community
1. 社區：具有共同經濟社會政治興趣的群組
2. 分析社群：更好的了解用戶，部分行為可能在個人層面南觀察到
3. 形成
    1. 至少兩個節點
    2. 共同興趣集合
    3. 節點之間產生交互
4. 形式
    1. emic顯式：可以直觀的在社交媒體上觀察到，例如用戶共同訂閱，加入小組
    2. etic隱式：其他非顯式的社區，例如把所有打電話到美國的個人，營運商將他們視為一個社區進行促銷
    3. 常常可以再大的顯式社區中找到小的隱式社區
5. 依照是否重疊，可以分成重疊社區或非重疊
6. 社區分析方向
    1. 社區發現：尤其是隱式社區
    2. 社區演變
    3. 社區評估
